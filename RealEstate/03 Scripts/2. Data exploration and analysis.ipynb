# 02. Exploring the dataset

```python
df.describe()
```

```python
#Resetting display settings
pd.options.display.max_rows=None
pd.options.display.max_columns=None
```

```python
#Grouping house price data by city 
df.groupby('city').agg({'price':['mean', 'min', 'max']})
```

```python
df.loc[df['price']==34000.0]
```

```python
#Average, minimum and maximum house prices per city
df.groupby('city').agg({'price':['mean', 'min', 'max'], 'avg_annual_income':['mean'], 'property tax rate':['mean']})
```

```python
#Average, minimum and maximum house prices per sqft per city
df.groupby('city').agg({'price_per_sqft':['mean', 'min', 'max'], 'avg_annual_income':['mean'], 'property tax rate':['mean']})
```

```python
#Average, minimum and maximum house prices per sqftlot per city
df.groupby('city').agg({'price_per_sqft_lot':['mean', 'min', 'max'], 'avg_annual_income':['mean'], 'property tax rate':['mean']})
```
